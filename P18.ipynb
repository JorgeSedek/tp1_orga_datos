P18) Queremos saber quién de los 2 actores le gusta más a la gente, si Guillermo Francella o Ricardo Darin, para eso queremos saber quien tiene en promedio mejor vote_average en movies.csv.?


```python
df_peliculas = pd.read_csv('/content/drive/MyDrive/orgaDatos/Movie dataset/movies.csv')
```

```python
df_actores = pd.read_csv('/content/drive/MyDrive/orgaDatos/Movie dataset/cast.csv')
```

```python
df_imdb_actores = pd.read_csv('/content/drive/MyDrive/orgaDatos/Movie dataset/imdb_actors.csv')
```

```python
df_peliculas['imdb_id']
```

```python
idmbs_francella = df_imdb_actores[df_imdb_actores['primaryName'].map(lambda  x: x == 'Guillermo Francella')]
idmbs_francella['imdb_id']  = idmbs_francella['knownForTitles'].astype('string').apply(lambda x: x.split(','))
idmbs_francella = idmbs_francella.explode('imdb_id')
imdbs_francella_pelis = idmbs_francella[['primaryName','imdb_id']]
imdbs_francella_pelis
```

```python
idmbs_darin = df_imdb_actores[df_imdb_actores['primaryName'].map(lambda  x: x == 'Ricardo Darín')]
idmbs_darin['imdb_id']  = idmbs_darin['knownForTitles'].astype('string').apply(lambda x: x.split(','))
idmbs_darin = idmbs_darin.explode('imdb_id')
imdbs_darin_pelis = idmbs_darin[['primaryName','imdb_id']]
imdbs_darin_pelis 
```

```python
actores_peliculas = df_actores[['id','name']]
```

```python
peliculas_francella = actores_peliculas[actores_peliculas['name'].map(lambda  x: x == 'Guillermo Francella')]
```

```python
peliculas_darin = actores_peliculas[actores_peliculas['name'].map(lambda  x: x == 'Ricardo Darín')]
```

```python
peliculas_darin['id'] = peliculas_darin['id'].astype('string')
peliculas_francella['id'] = peliculas_francella['id'].astype('string')
idmbs_rating = df_peliculas[['imdb_id','vote_average']]
```

```python
tabla_francella = pd.merge(peliculas_francella, peliculas_rating)
tabla_francella = pd.merge(tabla_francella, imdbs_francella_pelis, how = 'outer')
tabla_francella = tabla_francella[tabla_francella['vote_average'] > 0]
tabla_francella
```

```python
tabla_darin = pd.merge(peliculas_darin, peliculas_rating)
tabla_darin = pd.merge(tabla_darin, imdbs_francella_pelis, how = 'outer')
tabla_darin = tabla_darin[tabla_darin['vote_average'] > 0]
tabla_darin
```

```python
darin_media = tabla_darin.groupby('name').mean()
darin_media
```

```python
francella_media = tabla_francella.groupby('name').mean()
francella_media 
```

```python
pd.concat([francella_media,darin_media])
```

Francella tiene mayor promedio.
